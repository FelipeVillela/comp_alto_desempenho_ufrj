```python
import subprocess
import time
import matplotlib.pyplot as plt
```

```python
max_iterations = 100
max_size = 60000

sizes = []
results_c = []
results_fortran = []

def get_exec_time(cmd_line):
    start_time = time.time()
    
    # Executa o código com base na command line e aguarda a conclusão
    subprocess.run(cmd_line, shell=True, check=True)
    
    end_time = time.time()
    elapsed_time = end_time - start_time
    return elapsed_time

for i in range(max_iterations + 1):
    n = i * int((max_size/max_iterations))
    sizes.append(n)
    
    # Executa o programa em C e guarda o tempo
    results_c.append(
        get_exec_time(f"./c_executable {n} {n}")
    )
    
    # Executa o programa em Fortran e guarda o tempo
    results_fortran.append(
        get_exec_time(f"./fortran_executable {n} {n}")
    )
```

```python
# Plotagem dos resultados
plt.plot(sizes, results_c, label='C')
plt.plot(sizes, results_fortran, label='Fortran')

# Configurações do gráfico
plt.xlabel('Tamanho do Vetor (n)')
plt.ylabel('Tempo (s)')
plt.title('Tempo de Execução em Função do Tamanho do Vetor')
plt.legend()

# Exibição do gráfico
plt.grid(True)
plt.savefig(f'resultado.png')
plt.show()

```
